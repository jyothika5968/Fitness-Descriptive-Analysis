```python
import statistics
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("/content/gym_members_exercise_tracking_synthetic_data.csv")
df.info()
df.head()
df.tail()
```

```python
print(df.isnull().sum())
```

```python
print(df.duplicated().sum())
```

```python

# Display first 5 rows
print(df.head())

# Descriptive Analysis
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
print("Mean =", df['Calories_Burned'].mean())

print("Median =", df['Calories_Burned'].median())

print("Mode =")
print(df['Calories_Burned'].mode())

print("Variance =", df['Calories_Burned'].var())

print("Standard Deviation =", df['Calories_Burned'].std())

print("Minimum =", df['Calories_Burned'].min())

print("Maximum =", df['Calories_Burned'].max())

print("Range =", df['Calories_Burned'].max() - df['Calories_Burned'].min())
```

```python
print(df.columns)
calories_burned = df['Calories_Burned']
plt.hist(calories_burned, bins=10)
plt.xlabel('Calories Burned')
plt.ylabel('Frequency')
plt.title('Histogram of Calories Burned')
plt.show()
```

```python
# Create Pie Chart
plt.figure(figsize=(7,7))

workout_counts = df['Workout_Type'].value_counts()
plt.pie(workout_counts,labels=workout_counts.index,autopct='%1.1f%%')
plt.title("Workout Type Distribution")

plt.show()
# Scatter plot
```

```python
# Scatter Plot
plt.figure(figsize=(8,6))

plt.scatter(df['Session_Duration (hours)'] * 60,
            df['Calories_Burned'])

# Titles and labels
plt.title("Workout Duration vs Calories Burned")

plt.xlabel("Workout Duration (mins)")
plt.ylabel("Calories Burned")

plt.show()
```
